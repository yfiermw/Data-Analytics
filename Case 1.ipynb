```python
# right tail
p = 0.1 #Defining tails as top p% of returns (both positive and negative)
bts = 0.8 # Fraction of data to be ratined in each bootsrap sample
Nbts = 500 # Number of bootstrap samples
alpha = 0.9 # Significance level

x2 = np.sort(r)
x2_right = x2[int(np.round((1-p)*len(r))): ] #% Selecting top p% of returns
N = len(x2_right) #Number of returns selected as right tail
alpha_right = N/sum(np.log(x2_right/min(x2_right))) #Maximum-Likelihood estimate for right tail exponent

print("Right tail exponent:{:.3f}".format(alpha_right))

x_right = np.linspace(min(x2_right), max(x2_right), num=100)
y_right = alpha_right*(x_right/min(x2_right))**(-alpha_right-1)/min(x2_right) #Power law distribution

fig = plt.figure(figsize=plt.figaspect(0.25))
ax = fig.add_subplot(1, 2, 1)

b_right, a_right, ptchs_x = plt.hist(x2_right, bins=20, normed= True)


ax = fig.add_subplot(1, 2, 2)
ax.loglog(a_right[:-1],b_right, marker='o',linewidth=0, markersize=8,color = 'r')
ax.loglog(x_right, y_right,'g', linewidth=2, label='Powerlaw pdf' )

ax.set_title("RIght tail", fontsize=18)
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
##Right tail with bootsrap
p = 0.1 #Defining tails as top p% of returns (both positive and negative)
bts = 0.8 # Fraction of data to be ratined in each bootsrap sample
Nbts = 500 # Number of bootstrap samples
alpha = 0.9 # Significance level

alpha_right_bts = []

for i in range(Nbts):
    
    r_bts = np.random.permutation(r) #Random permutation of returns
    r_bts = r_bts[:int(np.round(bts*len(r_bts)))] # Bootstrapping bts of returns
    r_bts = np.sort(r_bts)
    r_right_bts = r_bts[int(np.round((1-p)*len(r_bts))):] #Selecting top p% of returns
    
    N_bts = len(r_right_bts) #Number of bootstrapped returns
    
    alpha_right_bts.append(N_bts/sum(np.log(r_right_bts/min(r_right_bts))))
    
alpha_right_bts = np.sort(alpha_right_bts)
print("Mean of bootstrap:{:.3f}".format(np.mean(alpha_right_bts)))

right_interval_lower = alpha_right_bts[int(np.round(0.5*(1-alpha)*Nbts))]
right_interval_upper = alpha_right_bts[int(np.round(0.5*(1+alpha)*Nbts))]

print("Right tail interval at 0.90 CL: [{:.3f}, {:.3f}]".format(right_interval_lower,  right_interval_upper))
```

```python
# left tail
p = 0.1 #Defining tails as top p% of returns (both positive and negative)
bts = 0.8 # Fraction of data to be ratined in each bootsrap sample
Nbts = 500 # Number of bootstrap samples
alpha = 0.9 # Significance level

x2 = np.sort(r)
x2_left = x2[0 : int(np.round(p*len(r)))] #% Selecting top p% of returns
x2_left = np.abs(x2_left) #% Converting negative returns to positive numbers
N = len(x2_left) #Number of returns selected as right tail
alpha_left = N/sum(np.log(x2_left/min(x2_left))) #Maximum-Likelihood estimate for right tail exponent

print("left tail exponent:{:.3f}".format(alpha_left))

x_left = np.linspace(min(x2_left), max(x2_left), num=100)
y_left = alpha_left*(x_left/min(x2_left))**(-alpha_left-1)/min(x2_left) #Power law distribution

fig = plt.figure(figsize=plt.figaspect(0.25))
ax = fig.add_subplot(1, 2, 1)

b_left, a_left, ptchs_xleft = plt.hist(x2_left, bins=20, normed= True)


ax = fig.add_subplot(1, 2, 2)
ax.loglog(a_left[:-1],b_left, marker='o',linewidth=0, markersize=8,color = 'r')
ax.loglog(x_left, y_left,'g', linewidth=2, label='Powerlaw pdf' )

ax.set_title("left tail", fontsize=18)
ax.legend(loc='best', frameon=False)
plt.show()

```
